```python
import os
from dotenv import load_dotenv
load_dotenv()
#Here, we are going to use gemini 1.5 pro to see its capabilities.
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
genai.configure(api_key=os.getenv("API_KEY"))

# for model in genai.list_models():
#     print(model.name)
```

```python
#since we want a multi-modal, we will use gemini-2.5-flash to generate text and image both
model = genai.GenerativeModel("gemini-2.5-pro")
response = model.generate_content("Write a story about a magic backpack.")
print(response.text)
```

```python
from IPython.display import Markdown, display

def to_markdown(text):
    display(Markdown(text))
to_markdown(response.text)
```

```python
import re
from IPython.display import Image, display
import requests

# Combine all text parts into one string
text_output = "".join(part.text for part in response.candidates[0].content.parts)

# Find the image URL using regex
match = re.search(r'https?://[^\s)]+', text_output)
if match:
    image_url = match.group(0)
    print("✅ Image URL found:", image_url)
else:
    print("❌ No image URL found in response.")
```

```python
if match:
    img_data = requests.get(image_url).content
    with open("futuristic_city.png", "wb") as f:
        f.write(img_data)
    display(Image(filename="futuristic_city.png"))
```

```python
img_data = response._result.candidates[0].content.parts[0].inline_data.data
with open("futuristic_city.png", "wb") as f:
    f.write(img_data)

display(Image(filename="futuristic_city.png"))
```
